```python
import numpy as np
from manapy.cuda.utils import VarClass

def custom_fun():
  origin_fun = np.asarray
  def custom_fun(*args):
    print("np.array called")
    return VarClass(origin_fun(*args))
  return custom_fun

np.asarray = custom_fun()
```

```python
input_array = np.array([5, 8])
obj = np.ndarray.__new__(VarClass, shape=input_array.shape,
                                 dtype=input_array.dtype,
                                 buffer=input_array,
                                 strides=input_array.strides,
                                 order=None)

obj 
```
