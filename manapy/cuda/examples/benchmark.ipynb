```python
import matplotlib.pyplot as plt
import csv
import numpy as np
# Read data from CSV file

root_path = '/home/aben-ham/Desktop/work/stage/manapy/manapy/cuda/example/benchmark_result'
data1 = np.genfromtxt(f'{root_path}/1.benchmark_result_1.csv', delimiter=' ', dtype=[('col1', 'U10'), ('col2', 'f8')], skip_header=1)
data2 = np.genfromtxt(f'{root_path}/1.benchmark_result_2.csv', delimiter=' ', dtype=[('col1', 'U10'), ('col2', 'f8')], skip_header=1)
x1 = [a[0] for a in data1][:10]
y1 = [a[1] for a in data1][:10]
x2 = [a[0] for a in data2][:10]
y2 = [a[1] for a in data2][:10]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x1, y1, 'o', label='Time')
plt.plot(x2, y2, 'x', label='Time')

# Formatting the plot
plt.xlabel('Parameters (a, b)')
plt.ylabel('Time')
plt.title('Time vs Parameters')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
